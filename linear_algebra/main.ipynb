Representing Systems of Equations as Matrices

Packages

```python
import numpy as np
import matplotlib.pyplot as plt
from utils import plot_lines
```

{−𝑥1+3𝑥2=7,3𝑥1+2𝑥2=1} to be represented in it's matrix form using numpy

```python
A = np.array([[-1, 3], 
              [3, 2]], dtype=np.dtype(float))

b = np.array([7,1], dtype=np.dtype(float))
print(f"Matrix A:\n{A}")
print(f"Vector b:\n{b}")
```

To get the dimensions

```python
print(f"Shape of A: {np.shape(A)}")
print("Shape of b:", b.shape)
```

To solve the equation using numpy function

```python
sol = np.linalg.solve(A, b)

print ("Solution", sol)
```

To get the determinant

```python
d = np.linalg.det(A)

print(f"Determinant of matrix: {d:.2f}")
```

To represent the system as a matrix

```python
A_system = np.hstack((A, b.reshape((2,1))))

print(A_system)
```

To graphically represent the equations

```python
plt.scatter(sol[0], sol[1], color='red', s=50, zorder=5)

plt.annotate(f"({sol[0]:.0f}, {sol[1]:.0f})",
                     (sol[0], sol[1]),
                     textcoords="offset points",
                     xytext=(12, 15),
                     ha='center')
plot_lines(A,b)
```
